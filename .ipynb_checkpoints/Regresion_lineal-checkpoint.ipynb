```python
import pandas as pd

bike_df = pd.read_csv('data/daily-bike-share.csv')
bike_df.head(5)
```

```python
bike_df['day'] = pd.DatetimeIndex(bike_df['dteday']).day
```

```python
bike_df.head(5)
```

```python
#Resumen de elementos especificos (descriptivos)
resumen_st = ['temp','atemp','hum','windspeed']
bike_df[resumen_st + ['rentals']].describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#Ver la visualización dentro del chunck
```

```python
label = bike_df['rentals']


#Se crean 2 filas 1 columna
fig, ax = plt.subplots(2, 1, figsize = (9,12))

# Plot del histograma  
ax[0].hist(label, bins=100)
ax[0].set_ylabel('Frequency')

# se añaden lineas de media, mediana.
ax[0].axvline(label.mean(), color='magenta', linestyle='dashed', linewidth=2)
ax[0].axvline(label.median(), color='cyan', linestyle='dashed', linewidth=2)

# Plot un boxplot  
ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

# titulo
fig.suptitle('Rental Distribution')

# print
fig.show()
```

```python
#histograma para cada variable numerica
for col in resumen_st:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = bike_df[col]
    feature.hist(bins=100, ax = ax)
    ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
    
```

```python
import numpy as np

#bar plot de cada variable categorica
categorical_var = ['season','mnth','holiday','weekday','workingday','weathersit', 'day']

for col in categorical_var:
    counts = bike_df[col].value_counts().sort_index()
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    counts.plot.bar(ax = ax, color='steelblue')
    ax.set_title(col + ' counts')
    ax.set_xlabel(col) 
    ax.set_ylabel("Frequency")
plt.show()
```

```python
#correlación
for col in resumen_st:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = bike_df[col]
    label = bike_df['rentals']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs ' + col + '- correlation: ' + str(correlation))
plt.show()
```

```python
#box plot
#Optimizar este grafico
for col in resumen_st:
    fig = plt.figure(figsize=(12, 12))
    ax = fig.gca()
    bike_df.boxplot(column = 'rentals', by = col, ax = ax)
    ax.set_title('Label by ' + col)
    ax.set_ylabel("Bike Rentals")
plt.show()
```

```python
#Etiquetas y caracteristicas
X, y = bike_df[['season','mnth', 'holiday','weekday','workingday','weathersit','temp', 
                'atemp', 'hum', 'windspeed']].values, bike_df['rentals'].values
print('Features:',X[:10], '\nLabels:', y[:10], sep='\n')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, y_train)
print (model)
```

```python
predictions = model.predict(X_test)
np.set_printoptions(suppress=True)
print('Predicted labels: ', np.round(predictions)[:10])
print('Actual labels   : ' ,y_test[:10])
```
